# In this Part 4 - We are seeing

```python
1. \A and \Z Vs ^ and $ - Sinlge Line and Multi Line of text/String
2. \b - word bound 
3. \W  \D \S 
```

```python
import re

text = """welcome to Alex Tech Projects
welcome to Regular Expressions"""

pattern = re.compile(r"\Awelcome",re.MULTILINE)
print(re.findall(pattern, text))

pattern = re.compile(r"^welcome",re.MULTILINE)
print(re.findall(pattern, text))
```

```python
import re

text = """Like Share and Comment
Subscribe and Comment"""

pattern = re.compile(r"Comment\Z",re.MULTILINE)
print(re.findall(pattern, text))

pattern = re.compile(r"Comment$",re.MULTILINE)
print(re.findall(pattern, text))
```

# word boundry

```python
import re

text = "wordplay word keyword 1word word1" 

print("Start boundary (\\bword):", re.findall(r"\bword", text))
print("End boundary (word\\b):", re.findall(r"word\b", text))
print("Exact word (\\bword\\b):", re.findall(r"\bword\b", text))
```

```python
import re
text = "He is eating, dancing and sleeping and doing doingbath"
pattern =re.compile(r"\w+ing\b")
print(re.findall(pattern, text))
```

```python
import re
text = "He is eating, dancing and sleeping and doing"
pattern =re.compile(r"\bis\b")
print(re.findall(pattern, text))
```

\W \D \S

```python
\W  Non-word characters (anything except letters, digits, and _)

\D  Non-digits (anything except 0-9)

\S  Non-whitespace (anything except space, tab, newline, etc.)
```

```python
import re

text = """welcome to Alex Tech Project \n
welcome to Regular Expressions"""

print("Non-word characters (\\W):")
print(re.findall(r"\W", text))

print("\nNon-digits (\\D):")
print(re.findall(r"\D", text))

print("\nNon-whitespace (\\S):")
print(re.findall(r"\S", text))
```
