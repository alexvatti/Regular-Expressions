# match , search , finditer

```
txt = "re is my first Python Project"
pattern = re.compile(r"""
                    Python # search for Python
                    """,re.VERBOSE)

print(bool(re.search(pattern,txt)))
if bool(re.search(pattern,txt)):
    m=re.search(pattern,txt)
    print(f"{m.string}\n{m.span()}\n{m.start()}\n{m.end()}\n{m.group()}\n")
```

```
txt = "re is my first Python Project"
pattern = re.compile(r"""
                    Python # search for Python
                    """,re.VERBOSE)
print(bool(re.match(pattern,txt)))
if bool(re.match(pattern,txt)):
    m=re.match(pattern,txt)
    print(f"{m.string}\n{m.span()}\n{m.start()}\n{m.end()}\n{m.group()}\n")
```

```
txt = "Python is my first Python Project"
pattern = re.compile(r"""
                    Python # search for Python
                    """,re.VERBOSE)

iter_list = re.finditer(pattern,txt)
for m in iter_list:
    print(f"{m.string}\n{m.span()}\n{m.start()}\n{m.end()}\n{m.group()}\n")
```

# findall , fullmatch

```
txt = "Python is my first Python Project"
pattern = re.compile(r"""
                    Python # search for Pyhton
                    """,re.VERBOSE)
print(re.findall(pattern,txt))
```

```
txt = "Python"
pattern = re.compile(r"""
                    Python # search for Pyhton
                    """,re.VERBOSE)
print(re.fullmatch(pattern,txt))
```

# Meta/ Special Character in Re

- [anyone] , .  |  (group) \w , \s, \d ,^ , $ , {min.max} 
- + one or more
- * zero or more
- ? zero or one

```
 [abc] . (movie|cinema) , \w \d \s ^ $ {min,max}
    
- + one or more [ greedy]
- * zero or more [greedy]
- ? zero or one
```

```
txt = "My mobile 1234567890 1234569870"
pattern = r"\d{10}"
re.findall(pattern,txt)
```

```
txt = "My mobile 1234567890 1234569870"
pattern = r"[0-9]{10}"
re.findall(pattern,txt)
```

```
txt = "my mobile 1234567890 1234569870"
pattern = r"^(my|My)"
re.findall(pattern,txt)
```

```
txt = "Python is my first Python Project"
pattern = r"[P|p]roject$"
re.findall(pattern,txt)
```
